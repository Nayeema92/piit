```python
import pandas as pd
```

```python
Patient = pd.read_excel(r"C:\analytics\data\Patient.xlsx")
Patient
```

```python
DetailsP = pd.read_excel(r"C:\analytics\data\DetailsP.xlsx")
DetailsP
```

```python
##inner join
pd.merge(Patient,DetailsP,on="ID", how="inner")
```

```python
##left join
pd.merge(Patient,DetailsP,on="ID",how="left")
```

```python
##right join
pd.merge(Patient,DetailsP,on="ID",how="right")
```

```python
##outer join
pd.merge(Patient,DetailsP,on="ID",how="outer")
```

```python
#cross join
Patient.assign(key=1).merge(DetailsP.assign(key=1))
```
