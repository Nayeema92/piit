```python
#What’s the difference between NumPy arrays and pandas Series?
#✅ NumPy Array
A NumPy array is:
 
A homogeneous data structure (all elements must be of the same type).
 
Only stores values (numbers, strings, etc.).
 
Does NOT have labels (only numeric indices)
import numpy as np
arr = np.array([10, 20, 30])
```

```python
###✅ pandas Series

A pandas Series is:
 
A labeled one-dimensional array.
 
Can store heterogeneous data (mixed types).
 
Has both:
 
values
 
index labels (customizable)
 
 
##import pandas as pd

s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
 
```

```python
#How do you plot multiple lines on the same chart in matplotlib?
```

```python
x = [1, 2, 3, 4]
 
plt.plot(x, [1, 4, 9, 16], label='Line 1')
plt.plot(x, [2, 3, 2, 3], label='Line 2')
 
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines in One Plot")
plt.show()
```
